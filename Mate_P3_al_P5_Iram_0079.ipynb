#### Problema 3
Problema 3: La resta y el despeje inverso
* Planteamiento:
* Buscamos un número desconocido: $x$."Al que se le resta siete": $x - 7$."
Da como resultado 18": $= 18$.
* Ecuación:$$x - 7 = 18$$
* Despeje y Solución:Para dejar a la $x$ sola, el 7 que está restando pasa al otro lado de la igualdad con la operación opuesta (sumando).$x = 18 + 7$
* Resultado: $x = 25$

```python
# Problema 3
# Un número x menos 7 es igual a 18
resultado = 18
resta = 7

# Despeje: x = 18 + 7
numero = resultado + resta
print(f"El número del Problema 3 es: {numero}")
```

#### Problema 4
Problema 4: Ecuación de dos pasos (Multiplicación y Suma)
* Planteamiento:El número desconocido es $n$."El triple de un número": $3n$."Más cuatro": $+ 4$."Es igual a 31": $= 31$.
* Ecuación:$$3n + 4 = 31$$
* Despeje y Solución:Primero movemos lo que está sumando o restando.
* El 4 pasa restando:$3n = 31 - 4$$3n = 27$
* Ahora, el 3 que está multiplicando a la $n$ pasa al otro lado dividiendo:$n = \frac{27}{3}$
* Resultado: $n = 9$

```python
#Problema 4
# 3n + 4 = 31
total = 31
suma = 4
multiplo = 3

# Despeje: n = (31 - 4) / 3
n = (total - suma) / multiplo
print(f"El número del Problema 4 es: {int(n)}")
```

#### Problema 5
Problema 5: Geometría y Perímetro
* Planteamiento:Ancho del terreno: $a$Largo del terreno (cinco metros más que el ancho): $a + 5$
* El perímetro de un rectángulo es la suma de sus cuatro lados: $Largo + Largo + Ancho + Ancho$.Suma total: $(a + 5) + (a + 5) + a + a = 50$
* Ecuación:$$4a + 10 = 50$$
* Despeje y Solución:Pasamos el 10 restando:$4a = 50 - 10$$4a = 40$
* Pasamos el 4 dividiendo:$a = \frac{40}{4}$$a = 10$
* Resultado:Ancho: 10 metros.Largo: $10 + 5 = 15$ metros.
* Comprobación: $15 + 15 + 10 + 10 = 50$ m.

```python
#Problema 5
# 4a + 10 = 50
perimetro = 50
constante_largo = 10 # (5+5 de los dos lados largos)

# Despeje de 'a' (ancho)
ancho = (perimetro - constante_largo) / 4
largo = ancho + 5

print(f"Las medidas del terreno son: Ancho = {ancho}m, Largo = {largo}m")
```
